```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set globally pseudo-random generators at a fixed value.
# Seed value
seed_value= 0

# 1. Set `PYTHONHASHSEED` environment variable at a fixed value
import os
os.environ['PYTHONHASHSEED']=str(seed_value)

# 2. Set `python` built-in pseudo-random generator at a fixed value
import random
random.seed(seed_value)

# 3. Set `numpy` pseudo-random generator at a fixed value
import numpy as np
np.random.seed(seed_value)
```

```python
# fixed parameter
n = 1000000
t = np.logspace(np.log10(10), np.log10(500),n)

# variable parameter
A = [1, 1, 1.5, 1.5]
d = [.004, .001, .002, .0015]
f = [3, 1, 2, 2.5]

# generate XY value pairs
x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)

# plot
plt.plot(x, y, 'k', linewidth=0.1)
plt.axis('off')
plt.show()
```

```python
# what are the elementary components of the plots above?
plt.plot(np.sin(t*f[0])*np.exp(-d[0]*t), linewidth=.5)
plt.show()
```

the multiplication of t*f makes the function goes faster as t increases

while the exponential function tapering down the curve

```python
# take away the exponential function
plt.plot(np.sin(t*f[0]), linewidth=.5)
plt.show()
```

```python
# multiply the exponential components.
# it will be reduced faster
plt.plot(np.sin(t*f[0])*np.exp(-d[0]*5*t), linewidth=.5)
plt.show()
```

```python
# generate new arts ~
# fixed parameter
n = 1000000
t = np.logspace(np.log10(10), np.log10(500),n)

# variable parameter
A = [1, 1, 1.5, 1.5]
d = [.004, .001, .002, .0015]
f = [3, 1, 2, 2.5]

# generate XY value pairs
def generate_xy(A, d, f, t):
    x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
    y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)
    return x, y

# Set seed for reproducibility
#np.random.seed(0) # run on the first cell of this notebook

# Number of subplots
num_rows = 4
num_cols = 4
num_plots = num_rows * num_cols

# Create subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(10, 8), dpi=600)

# Flatten the axs array
axs = axs.flatten()

# Loop through each subplot
for i, ax in enumerate(axs):
    # Generate random parameters within a specified range
    random_A = np.round(np.random.uniform(0.5, 2, size=4), 2)
    random_d = np.round(np.random.uniform(0.0005, 0.002, size=4), 2)
    random_f = np.round(np.random.uniform(1, 4, size=4), 2)

    # Generate XY value pairs
    x, y = generate_xy(random_A, random_d, random_f, t)

    # Plot
    ax.plot(x, y, 'k', linewidth=0.1)
    ax.axis('off')
    ax.set_title(f"Plot {i+1}")

    # Annotate with parameter values
    ax.text(0.1, 0.9, f"A: {random_A}\nd: {random_d}\nf: {random_f}", transform=ax.transAxes, fontsize=6, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))


# Adjust layout
plt.tight_layout()
plt.show()
```
